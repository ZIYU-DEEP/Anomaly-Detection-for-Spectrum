```python
normal_1 = normal_data[:-1, :]
normal_2 = normal_data[1:, :]
abnormal_1 = abnormal_data[:-1, :]
abnormal_2 = abnormal_data[1:, :]
```

```python
normal_corrs = []

for i in range(len(normal_1)):
    normal_corrs.append(correlate(normal_1[i], normal_2[i])[0])

normal_corrs = np.array(normal_corrs)

abnormal_corrs = []

for i in range(len(abnormal_1)):
    abnormal_corrs.append(correlate(abnormal_1[i], abnormal_2[i])[0])

abnormal_corrs = np.array(abnormal_corrs)
```

```python
sea.distplot(normal_corrs)
```

```python
sea.distplot(abnormal_corrs)
```

```python
th = np.quantile(normal_corrs, 0.90)
scores = np.where(abnormal_corrs > th, 1, 0).reshape(-1)
scores.sum() / scores.size
```

```python
th = np.quantile(normal_corrs, 0.95)
scores = np.where(abnormal_corrs > th, 1, 0).reshape(-1)
scores.sum() / scores.size
```
